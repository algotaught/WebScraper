```python
import pandas as pd
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
website_url='https://www.flipkart.com'
base_url='https://www.flipkart.com/search?q='
```

```python
results_page=base_url+'iphone'
```

# getting the url to access the product page

```python
source=requests.get(results_page).content
soup=BeautifulSoup(source,'lxml')
soup
```

```python
type(soup)
```

```python
product=soup.find('a',class_='_31qSD5').get('href')
```

```python
product
```

```python
product_url=website_url+product
```

```python
product_url
```

# getting the url of the page which holds all the reviews

```python
source=requests.get(product_url).content
soup=BeautifulSoup(source,'lxml')
reviews_url=soup.find('div',class_='swINJg _3nrCtb').parent.get('href')
reviews_url=website_url+reviews_url
reviews_url
```

# Extracting the reviews

```python
source=requests.get(reviews_url).content
soup=BeautifulSoup(source,'lxml')
```

```python
all_reviews=soup.find_all('div',class_='_3gijNv col-12-12')
```

```python
reviews=[]
for review in all_reviews:
    
#extracting all the names one by one and storing them in name variable.
    try:
        if (review.find('p',class_='_3LYOAd _3sxSiS'))!=None:
            name=review.find('p',class_='_3LYOAd _3sxSiS').text
        else:
            continue
    except:
        name='not available'
        
#extracting all the comments one by one and storing them in comment variable        
    try:    
        if (review.find('div',class_='qwjRop'))!=None:
            comment=review.find('div',class_='qwjRop').div.div.text
        else:
            continue
    except:
        comment='not available'
        
#extracting all the ratings one by one and storing them in rating variable
    try:
        if (review.find('div',class_='hGSR34 E_uFuv'))!=None:
            rating=review.find('div',class_='hGSR34 E_uFuv').text
        else:
            continue
    except:
        rating='Rating not available'
#storing all the variables in a dictionary format
    try:
        reviews_dict={'Name':name,'Comment':comment,'Rating':rating}
    except:
        print('error creating the dictionary')
    
#storing the different dictionaries containing the review information into one single list. 
#So our output will be a list of different dictionaries.
    try:
        reviews.append(reviews_dict)
    except:
        print('Error creating the list of dictionaries')

```

```python
reviews
```

# Creating pandas DataFrame

```python
reviews_dataframe=pd.DataFrame([i for i in reviews])
```
